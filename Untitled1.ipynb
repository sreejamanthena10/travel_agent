### Create a sample PDF file for the vector database

For demonstration purposes, I will create a dummy PDF file. In a real scenario, you would use your actual hotel booking PDF(s).


```python
# Install necessary library to create PDF
!pip install -q reportlab

from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter

def create_dummy_pdf(filename="hotel_booking_details.pdf"):
    c = canvas.Canvas(filename, pagesize=letter)
    c.drawString(100, 750, "Hotel Booking Confirmation")
    c.drawString(100, 730, "Booking ID: XYZ789")
    c.drawString(100, 710, "Guest Name: John Doe")
    c.drawString(100, 690, "Hotel Name: Grand Imperial Hotel")
    c.drawString(100, 670, "Check-in: 2024-08-15")
    c.drawString(100, 650, "Check-out: 2024-08-20")
    c.drawString(100, 630, "Room Type: Deluxe King")
    c.drawString(100, 610, "Total Price: $1200")
    c.drawString(100, 590, "Confirmation Email: john.doe@example.com")
    c.save()
    print(f"Dummy PDF '{filename}' created.")

create_dummy_pdf()
```

### Create the Vector Database (`vector_db`)

Now, let's load the PDF, split it into manageable chunks, embed those chunks using a Google Generative AI embedding model, and then store them in a FAISS vector database. This `vector_db` will then be used by the `retriever_tool`.

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS

# Load the PDF document
loader = PyPDFLoader("hotel_booking_details.pdf")
docs = loader.load()

# Split the document into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(docs)

# Initialize the embedding model with a supported model
embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2")

# Create a FAISS vector database from the document chunks and embeddings
vector_db = FAISS.from_documents(splits, embeddings)

print("Vector database 'vector_db' created successfully.")
```

```python
import google.generativeai as genai

print("Listing all available Generative AI models:")
for m in genai.list_models():
    if "embedContent" in m.supported_generation_methods:
        print(f"Model: {m.name}, Description: {m.description}")
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_core.tools import create_retriever_tool

# Tool 1: Web Search (For real-time info)
search_tool = DuckDuckGoSearchRun()

# Tool 2: PDF Retriever (For your specific booking details)
# Note: This assumes you have your 'vector_db' already created from Phase 1
retriever_tool = create_retriever_tool(
    vector_db.as_retriever(),
    "hotel_booking_lookup",
    "Use this tool when answering questions about the user's specific hotel, booking dates, or confirmation details."
)

# Combine them into the tool list
tools = [search_tool, retriever_tool]
```

```python
from langchainhub import Client
from langchain_core.agents import AgentExecutor # Corrected import path for AgentExecutor
from langchain.agents.react import create_react_agent
from langchain_google_genai import ChatGoogleGenerativeAI

# 1. Initialize Gemini (using the 2026 stable version)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

# 2. Get the "Instructions" from LangChain Hub
prompt = Client().pull("hwchase17/react")

# 3. Create the Agent
agent = create_react_agent(llm, tools, prompt)

# 4. Create the Executor with Error Handling
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,           # This lets you see the agent "Thinking"
    handle_parsing_errors=True # Guardrail to prevent crashes
)
```

```python
result = agent_executor.invoke({"input": "What are my booking details?"})
print(result["output"])
```
